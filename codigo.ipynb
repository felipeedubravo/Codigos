```python

#2.Escreva um programa em Python que receba um número inteiro e diga se ele é par ou ímpar.
n = int(input("Digite um número inteiro: "))
if n % 2 == 0:
    print(f"O número {n} é par.")
else:
    print(f"O número {n} é ímpar.")

#3. Faça um Programa que verifique o estado civil de uma pessoa. Se a letra digitada é "C" (Casado), "S" (Solteiro), "D" (Divorciado), "V" (Viúvo) ou "O" (outros). Conforme a letra escrita pelo usuário seu programa deve escrever o estado civil, exemplo:
"""Usuário digita: C
Seu programa deve responder:
C - Casado
S - Solteiro
D - Divorciado
V - Viúvo
O - Outros
E assim sucessivamente"""

estado_civil = input("Digite o estado civil (C, S, D, V, O): ").upper()

if estado_civil == "C":
    print("C - Casado")
elif estado_civil == "S":
    print("S - Solteiro")
elif estado_civil == "D":
    print("D - Divorciado")
elif estado_civil == "V":
    print("V - Viúvo")
elif estado_civil == "O":
    print("O - Outros")
else:
    print("Estado civil inválido.")
```

```python
#1.Faça um programa que peça dois números ao usuário e mostre qual o maior e qual o menor número.

n1 = int(input("Digite o primeiro número: "))
n2 = int(input("Digite o segundo número: "))

if n1 > n2:
    print(f"O maior número é {n1} e o menor é {n2}")
else:
    print(f"O maior número é {n2} e o menor é {n1}")
```
