```python
# Escreva um programa em Python que receba um número inteiro e diga se ele é par ou ímpar.

# Recebe um número inteiro do usuário
n = int(input("Digite um número inteiro: "))

# Verifica se o número é par ou ímpar
if n % 2 == 0:
# se o número for par, imprime que é par
    print(f"O número {n} é par.")
    
# se o número não for par, então é ímpar
else:
    print(f"O número {n} é ímpar.")
```

```python
# Exemplo prático: Uma loja dá frete grátis se a compra for maior que R$ 100 OU se o cliente for VIP.
valor_compra = 50
cliente_vip = True

# Se a compra for alta OU se for VIP, ganha o frete
if valor_compra > 100 or cliente_vip == True:

# se a compra for alta OU se for VIP, imprime que ganhou frete grátis
    print("Você ganhou frete grátis!")
    
# se a compra for baixa E não for VIP, o frete será cobrado
else:
    print("O frete será cobrado.")
```

```python
# Faça um programa que peça dois números ao usuário e mostre qual o maior e qual o menor número.

# dois números ao usuário e mostre qual o maior e qual o menor número.
n1 = int(input("Digite o primeiro número: "))
n2 = int(input("Digite o segundo número: "))

# verifica qual é o maior e o menor número
if n1 > n2:
    print(f"O maior número é {n1} e o menor é {n2}")

# se n2 for maior que n1, imprime n2 como maior e n1 como menor
elif n2 > n1:
    print(f"O maior número é {n2} e o menor é {n1}")
    
# se n1 e n2 forem iguais, imprime que os dois números são iguais
else:
    print("Os dois números são iguais.")
```

```python
# Faça um Programa que verifique o estado civil de uma pessoa. Se a letra digitada é "C" (Casado), "S" (Solteiro), "D" (Divorciado), "V" (Viúvo) ou "O" (outros). Conforme a letra escrita pelo usuário seu programa deve escrever o estado civil, exemplo:

"""Usuário digita: C
Seu programa deve responder:
C - Casado
S - Solteiro
D - Divorciado
V - Viúvo
O - Outros
E assim sucessivamente"""

# Recebe o estado civil do usuário
estado_civil = input("Digite o estado civil (C, S, D, V, O): ").upper()

# Verifica o estado civil e imprime a resposta correspondente
if estado_civil == "C":
    print("C - Casado")
elif estado_civil == "S":
    print("S - Solteiro")
elif estado_civil == "D":
    print("D - Divorciado")
elif estado_civil == "V":
    print("V - Viúvo")
elif estado_civil == "O":
    print("O - Outros")

# Se o usuário digitar uma letra inválida, imprime que o estado civil é inválido
else:
    print("Estado civil inválido.")
```
