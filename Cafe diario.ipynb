```python

# import bibliotecas que voce vai usar

# Tempo Exeplo - time.sleep(3)
import time

# Para automatização 
import pyautogui

# pausa a execução do programa por 3 segundo
time.sleep(3)

# define um atraso global de 1.5 segundos após a execução de cada comando
pyautogui.PAUSE = 1.5

# Controle dos tequado, escrever, click do mause
pyautogui.press("enter")
pyautogui.write("chrome")
pyautogui.click(x=-1092, y=388) 





#==============================================================================
def cadastrar_produto():
    produto = input('Digite o nome do produto que deseja cadastrar: ')
# função de string usada para converter todos os caracteres em letras minúsculas
    produto = produto.casefold()
# limpar o texto função é remover espaços em branco (incluindo espaços, tabulações e quebras de linha) 
    produto = produto.strip()
# finalizar uma função quando o comando 'return' foi chamada retornar o resultado 
    return produto
#===============================================================================

# Chamando a função e guardando o retorno em uma variável
produto = cadastrar_produto()
print(produto)



```
