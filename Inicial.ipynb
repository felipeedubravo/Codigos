### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pandas
    

# Passo 4: Calcular os indicadores (faturamento e qualidade de produtos vendidos) a partir do arquivo baixado
    # abrir a base dedados
Caminho = r"C:\Users\Usuario\Downloads\Vendas - Dez.xlsx"
tabela = pandas.read_excel(Caminho)

# ver as informações de faturamento e qualidade de produtos vendidos
display(tabela)
# somar o faturamento ede todas os produtos vendidos = somar a coluna "Valor Final"
faturamento = tabela["Valor Final"].sum()
# somar a quantidade de produtos vendidos = somar a coluna "Quantidade"
quantidade = tabela["Quantidade"].sum()

print(faturamento)
print(quantidade)   
```
